```python
def load_labels(label_path):
    lines = []
    with open(label_path, "r") as f:
        for line in f.readlines():
            line = line.replace("\n", "")
            line = [float(elem) for elem in line.split(" ")]
            lines.append(line)
    return lines
```

```python
def draw_polygon(image, annotation, width, height):
    points = annotation[1:]
    polygon_points = []
    for i in range(0, len(points), 2):
        x = int(points[i] * width)
        y = int(points[i + 1] * height)
        polygon_points.append((x, y))

    cv2.polylines(image, [np.array(polygon_points)], isClosed=False, color=(255, 0, 0), thickness=2)

    # Ajouter l'étiquette de classe près du premier point
    cv2.putText(image, "rooftop", polygon_points[0], cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
```

## From polygone to mask

```python
def generate_mask(mode="train"):
    mask_folder = os.path.join(base_path, mode, "masks")
    if not os.path.exists(mask_folder):
        os.makedirs(mask_folder)

    images_path = os.path.join(base_path, mode, "images")
    labels_path = os.path.join(base_path, mode, "labels")
    
    for file in os.listdir(images_path):
        base_name = ".".join(file.split(".")[:-1])

        output_mask_path = os.path.join(mask_folder, f"{base_name}.png")

        if os.path.exists(output_mask_path):
            continue
                                        
        image = Image.open(os.path.join(images_path, file))
        width, height = image.size

        mask = np.zeros((height, width), dtype=np.uint8)

        label_file = os.path.join(labels_path, f"{base_name}.txt")
        
        polygons = load_labels(label_file)

        for idx, poly in enumerate(polygons):
            keypoints = poly[1:]  # first is class

            pts = [(int(keypoints[i]*width), int(keypoints[i+1]*height)) for i in range(0, len(keypoints), 2)]
            if len(pts) >= 3:
                cv2.fillPoly(mask, [np.array(pts, dtype=np.int32)], 255)
            else:
                print(f"Warning. Got less then 3 points for {base_name}")
        mask_img = Image.fromarray(mask)
        mask_img.save(output_mask_path)
    print("Fin ", mode)
```

### Option 1: linear interpolation (resampling)

```python
def normalize_rooftop(mode="train", num_points=15):
    
    new_labels_path = os.path.join(base_path, mode, f"labels_resampled_{num_points}")
    if not os.path.exists(new_labels_path):
        os.makedirs(new_labels_path)
        
    labels_path = os.path.join(base_path, mode, "labels")
    
    for file in os.listdir(labels_path):
        file_path = os.path.join(labels_path, file)
        
        output_file_path = os.path.join(new_labels_path, file)
        if os.path.exists(output_file_path):
            continue
            
        with open(file_path, "r") as input_file:
            with open(output_file_path, "w") as output_file:
                for line in input_file:
                    parts = line.strip().split()
                    class_idx = parts[0]
                    points = list(map(float, parts[1:]))  # list of x1, y1, x2, y2, ..., xN, yN
                    points_array = np.array(points).reshape(-1, 2)  # array of [[x1 y1], [x2 y2], ..., [xN, yN]]
            
                    if len(points_array) < 2:
                        continue  # trop peu de points pour interpoler
            
                    distances = np.cumsum(np.linalg.norm(np.diff(points_array, axis=0), axis=1))
                    distances = np.insert(distances, 0, 0)
                        
                    x_interp = scipy.interpolate.interp1d(distances, points_array[:, 0], kind='linear')
                    y_interp = scipy.interpolate.interp1d(distances, points_array[:, 1], kind='linear')
                    new_distances = np.linspace(0, distances[-1], num_points)
                    new_points = np.column_stack([x_interp(new_distances), y_interp(new_distances)])
            
                    # Format final
                    normalized_line = [class_idx] + list(new_points.flatten())
                    output_file.write(' '.join(map(str, normalized_line)) + '\n')
    print(f"OK for {mode}")
```

```python
normalize_rooftop(mode="train", num_points=mean_pts_per_polygone)
normalize_rooftop(mode="test", num_points=mean_pts_per_polygone)
normalize_rooftop(mode="valid", num_points=mean_pts_per_polygone)
```

# Prepare Data

```python
class CustomDataset(Dataset):
    def __init__(self, subset, mode="masks", mean_pts=7, transforms=None):
        self.subset = subset
        self.mode = mode
        self.mean_pts = mean_pts
        self.image_dir = os.path.join(base_path, self.subset, "images")
        self.mask_dir = os.path.join(base_path, self.subset, "masks")
        if not os.path.exists(self.mask_dir):
            warnings.warn(f"Error, directory {self.mask_dir} does not exists")
        
        self.labels_polygones = os.path.join(base_path, self.subset, f"labels_resampled_{self.mean_pts}")
        if not os.path.exists(self.labels_polygones):
            warnings.warn(f"Error, directory {self.labels_polygones} does not exists")

        self.image_files = [f for f in os.listdir(self.image_dir)]
        self.transforms = transforms

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        sample_name = self.image_files[idx]
        
        img_name = os.path.join(self.image_dir, sample_name)
        img = read_image(img_name).float()
        img = img/255.0

        height, width = img.shape[1], img.shape[2]
        
        target = {}
        
        if self.mode == "masks":
            mask_name = os.path.join(self.mask_dir, sample_name.replace(".jpg", ".png"))
            mask = read_image(mask_name)[0]
            mask = mask == 255

            num_objs = 1 if mask.any() else 0
            
            if num_objs == 0:
                boxes = torch.zeros((0, 4), dtype=torch.float32)
                masks = torch.zeros((0, height, width), dtype=torch.uint8)
                labels = torch.zeros((0,), dtype=torch.int64)
                iscrowd = torch.zeros((0, ), dtype=torch.int64)
                area = torch.zeros((0,), dtype=torch.float32)
            else:
                masks = mask.unsqueeze(0).to(torch.uint8)  # (1, H, W)  # masks = mask == obj_ids[:, None, None] if more than 1 kind of object is in the image

                boxes = masks_to_boxes(masks)
                labels = torch.ones((num_objs,), dtype=torch.int64)
                iscrowd = torch.zeros((num_objs,), dtype=torch.int64)
                area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0])
            
            target["masks"] = tv_tensors.Mask(masks)
            target["iscrowd"] = iscrowd
            target["area"] = area
            
        elif self.mode == "polygones":
            file_label = os.path.join(self.labels_polygones, sample_name.replace(".jpg", ".txt"))
            with open(file_label, "r") as f:
                boxes = []
                labels = []
                lines = f.readlines()
                if len(lines) != 0:
                    for line in lines:
                        parts = line.strip().split()
                        points = list(map(float, parts[1:]))
                        points_array = np.array(points).reshape(-1, 2)
    
                        points_array[:, 0] *= width
                        points_array[:, 1] *= height
    
                        x_min = np.min(points_array[:, 0])
                        y_min = np.min(points_array[:, 1])
                        x_max = np.max(points_array[:, 0])
                        y_max = np.max(points_array[:, 1])
                        boxes.append([x_min, y_min, x_max, y_max])
                        labels.append(1)  # it's 0 but Detection model automaticaly use 0 as the background
                else:
                    boxes = torch.zeros((0, 4), dtype=torch.float32)
                    labels = torch.zeros((0,), dtype=torch.int64)
                
            boxes = torch.tensor(boxes, dtype=torch.float32)
            labels = torch.tensor(labels, dtype=torch.int64)
        
        target["boxes"] = tv_tensors.BoundingBoxes(boxes, format="XYXY", canvas_size=F.get_size(img))
        target["labels"] = labels
        target["image_id"] = torch.tensor([idx])
        
        img = tv_tensors.Image(img)

        if self.transforms is not None:
            img, target = self.transforms(img, target)

        return img, target
```

# Prepare model

```python
def get_model(num_classes, backbone="resnet50", mode="masks"):
    if mode == "masks":
        if backbone == "resnet50":
            model = torchvision.models.detection.maskrcnn_resnet50_fpn(weights="DEFAULT")
            
        elif backbone == "mobilenetv2":
            backbone = torchvision.models.mobilenet_v2(weights="DEFAULT").features
            backbone.out_channels = 1280
            anchor_generator = torchvision.models.detection.anchor_utils.AnchorGenerator(
                sizes=((32, 64, 128, 256, 512),),
                aspect_ratios=((0.5, 1.0, 2.0),) * 5
            )
            roi_pooler = torchvision.ops.MultiScaleRoIAlign(
                featmap_names=['0'], output_size=7, sampling_ratio=2
            )
            model = detection.MaskRCNN(
                backbone,
                num_classes=91,
                rpn_anchor_generator=anchor_generator,
                box_roi_pool=roi_pooler
            )
        
        # get number of input features for the classifier
        in_features = model.roi_heads.box_predictor.cls_score.in_features
        # replace the pre-trained head with a new one
        model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)
    
        # now get the number of input features for the mask classifier
        in_features_mask = model.roi_heads.mask_predictor.conv5_mask.in_channels
        hidden_layer = 256
        # and replace the mask predictor with a new one
        model.roi_heads.mask_predictor = MaskRCNNPredictor(
            in_features_mask,
            hidden_layer,
            num_classes
        )
    elif mode == "polygones":
        model = detection.fasterrcnn_resnet50_fpn_v2(weights="DEFAULT")

        in_features = model.roi_heads.box_predictor.cls_score.in_features
        model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)
    
    return model

num_classes = 1 + 1 # background + class, ie rooftop
model = get_model(num_classes, backbone="mobilenetv2", mode="masks").to(device)
```

```python
def check_trainable_params(model, verbose=False):
    """Vérifie et affiche les paramètres entraînables"""
    total_params = 0
    trainable_params = 0
    
    for name, param in model.named_parameters():
        total_params += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
            if verbose:
                print(f"✓ {name}: {param.numel():,} params")
        else:
            if verbose:
                print(f"✗ {name}: {param.numel():,} params (frozen)")
    
    print(f"Params: {trainable_params:,}/{total_params:,} trainable ({100*trainable_params/total_params:.1f}%)")
    return trainable_params, total_params
```

```python
from torchvision.transforms import v2 as T

def get_transform(train):
    transforms = []
    if train:
        transforms.append(T.RandomHorizontalFlip(0.5))
    transforms.append(T.ToDtype(torch.float, scale=True))
    transforms.append(T.ToPureTensor())
    return T.Compose(transforms)
```

```python
def setup_training_stage(model, stage="stage1"):
    """Configure le modèle pour chaque stage d'entraînement"""
    
    if stage == "stage1":
        print("\n🚀 STAGE 1: Training heads only (backbone frozen)")
        # Freezer tout le backbone
        for param in model.backbone.parameters():
            param.requires_grad = False
            
        # S'assurer que les têtes sont défreezées
        for param in model.roi_heads.parameters():
            param.requires_grad = True
        for param in model.rpn.parameters():
            param.requires_grad = True
            
        return {
            'lr': 0.005,
            'epochs': 5,
            'description': 'Heads only - High LR'
        }
        
    elif stage == "stage2":
        print("\n🔥 STAGE 2: Training heads + late backbone layers")
        # Défreeze tout le backbone
        for param in model.backbone.parameters():
            param.requires_grad = True
            
        # Mais freeze les premières couches
        backbone_features = model.backbone.body
        for i in range(10):  # Freeze les 10 premières couches
            if i < len(backbone_features):
                for param in backbone_features.parameters():
                    param.requires_grad = False
        
        return {
            'lr': 0.002,
            'epochs': 3,
            'description': 'Heads + late backbone - Medium LR'
        }
        
    elif stage == "stage3":
        print("\n🎯 STAGE 3: Fine-tuning everything")
        # Tout défreezé
        for param in model.parameters():
            param.requires_grad = True
            
        return {
            'lr': 0.001,
            'epochs': 2,
            'description': 'Full model - Low LR'
        }
    
    else:
        raise ValueError(f"Stage {stage} not recognized. Use 'stage1', 'stage2', or 'stage3'")
```

# Train the model

```python
writer = SummaryWriter(log_dir='runs/rooftop')

def train_one_epoch(model, data_loader, device, optimizer, print_freq, epoch):
    model.train()
    metric_logger = utils.MetricLogger(delimiter="  ")
    metric_logger.add_meter("lr", utils.SmoothedValue(window_size=1, fmt="{value:.6f}"))
    header = f"Training Epoch {epoch}:"
    model.to(device)

    with tqdm(data_loader, desc=header) as tq:
        for i, (images, targets) in enumerate(tq):
            images = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            loss_dict = model(images, targets)
            losses = sum(loss for loss in loss_dict.values())

            loss_dict_reduced = utils.reduce_dict(loss_dict)
            losses_reduced = sum(loss for loss in loss_dict_reduced.values())

            loss_value = losses_reduced.item()

            optimizer.zero_grad()
            losses_reduced.backward()
            optimizer.step()

            metric_logger.update(loss=loss_value, **loss_dict_reduced)
            metric_logger.update(lr=optimizer.param_groups[0]["lr"])

            # Update tqdm postfix to display loss on the progress bar
            tq.set_postfix(loss=losses.item(), lr=optimizer.param_groups[0]["lr"])

            # Log losses to TensorBoard
            writer.add_scalar('Loss/train', losses.item(), epoch * len(data_loader) + i)
            for k, v in loss_dict.items():
                writer.add_scalar(f'Loss/train_{k}', v.item(), epoch * len(data_loader) + i)

    print(f"Average Loss: {metric_logger.meters['loss'].global_avg:.4f}")
    writer.add_scalar('Loss/avg_train', metric_logger.meters['loss'].global_avg, epoch)
    return metric_logger.meters['loss'].global_avg
```

```python
def evaluate(model, data_loader, device, epoch):
    model.eval()
    metric = MeanAveragePrecision()
    header = "Validation:"
    total_steps = len(data_loader)

    with torch.no_grad(), tqdm(total=total_steps, desc=header) as progress_bar:
        for i, (images, targets) in enumerate(data_loader):
            images = [img.to(device) for img in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
            outputs = model(images)

            # Convert outputs for torchmetrics
            preds = [
                {"masks": out["masks"], "boxes": out["boxes"], "scores": out["scores"], "labels": out["labels"]}
                for out in outputs
            ]
            targs = [
                {"masks": tgt["masks"], "labels": tgt["labels"], "boxes": tgt["boxes"]}
                for tgt in targets
            ]

            # Update metric for mAP calculation
            metric.update(preds, targs)
            progress_bar.update(1)

    results = metric.compute()

    # Log mAP to TensorBoard
    for k, v in results.items():
        if v.numel() == 1:  # Single element tensor
            writer.add_scalar(f'mAP/{k}', v.item(), epoch)
        else:  # Multi-element tensor, log each element separately
            for idx, value in enumerate(v):
                writer.add_scalar(f'mAP/{k}_{idx}', value.item(), epoch)
    return results
```

```python
def progressive_training(model, device, save_dir):
    """Lance l'entraînement progressif complet"""
    
    print("=" * 60)
    print("🎯 ENTRAÎNEMENT PROGRESSIF MASK R-CNN + RESNET")
    print("=" * 60)

    best_map = -float('inf') # Training loop
    stages = ['stage1', 'stage2', 'stage3']
    
    for stage in stages:
        # Configuration du stage
        config = setup_training_stage(model, stage)
        check_trainable_params(model)
        
        # Optimizer pour ce stage
        trainable_params = [p for p in model.parameters() if p.requires_grad]
        optimizer = torch.optim.SGD(
            trainable_params, 
            lr=config['lr'], 
            momentum=0.9,
            weight_decay=0.005
        )
        
        # Scheduler
        lr_scheduler = torch.optim.lr_scheduler.StepLR(
            optimizer, step_size=10, gamma=0.8
        )
        
        print(f"Optimizer: {len(trainable_params)} paramètres, LR={config['lr']}")
        print(f"Training: {config['epochs']} epochs")
        print("-" * 40)
        
        # Entraînement pour ce stage
        stage_losses = []
        for epoch in range(config['epochs']):
            epoch_loss = train_one_epoch(model, train_data_loader, device, optimizer, print_freq=10, epoch=epoch)
            
            stage_losses.append(epoch_loss)
            lr_scheduler.step()
                
            results = evaluate(model, val_data_loader, device, epoch)

            print(results)

            current_map = results['map'].item()
            if current_map > best_map:
                best_map = current_map
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': model.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'best_map': best_map,
                }, os.path.join(save_dir, f'best_model_checkpoint_epoch_{stage}_{epoch}.pth'))
                best_stage = stage
                best_epoch_model = epoch
                print(f"\tModel saved at {best_stage} {best_epoch_model} with mAP: {best_map}")
                
                    
        
        avg_stage_loss = sum(stage_losses) / len(stage_losses)
        print(f"🏁 {stage.upper()} terminé | Loss moyenne: {avg_stage_loss:.4f}")
    
    print("✨ ENTRAÎNEMENT PROGRESSIF TERMINÉ !")
    return model, best_stage, best_epoch_model
```
